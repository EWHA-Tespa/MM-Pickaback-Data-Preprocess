```python
import os
import random
import shutil

# 원본 이미지 저장 경로 바꿔야함!!
src_dir = "/data/youlee/data_library/cub_200_2011/images"

# 복사할 대상 경로 바꿔야함!!
dest_train_dir = "/data/youlee/data_library/cub_200_2011/image/train"
dest_test_dir = "/data/youlee/data_library/cub_200_2011/image/test"


os.makedirs(dest_train_dir, exist_ok=True)
os.makedirs(dest_test_dir, exist_ok=True)

folder_counts = {}
for label_folder in os.listdir(src_dir):
    folder_path = os.path.join(src_dir, label_folder)
    if os.path.isdir(folder_path):
        image_files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]
        folder_counts[label_folder] = len(image_files)

filtered_counts = {label: count for label, count in folder_counts.items() if count == 60}
labels_sorted = sorted(filtered_counts.keys())

group_size = 5
groups = [labels_sorted[i:i+group_size] for i in range(0, len(labels_sorted), group_size)]
groups = [group for group in groups if len(group) == group_size]

for group_index, group in enumerate(groups, start=1):
    group_train_dir = os.path.join(dest_train_dir, f"group{group_index}")
    group_test_dir = os.path.join(dest_test_dir, f"group{group_index}")
    os.makedirs(group_train_dir, exist_ok=True)
    os.makedirs(group_test_dir, exist_ok=True)
    
    for label in group:
        src_label_dir = os.path.join(src_dir, label)
        dest_train_label_dir = os.path.join(group_train_dir, label)
        dest_test_label_dir = os.path.join(group_test_dir, label)
        os.makedirs(dest_train_label_dir, exist_ok=True)
        os.makedirs(dest_test_label_dir, exist_ok=True)
        
        images = [f for f in os.listdir(src_label_dir) if os.path.isfile(os.path.join(src_label_dir, f))]
        random.shuffle(images)
    
        split_index = int(0.8 * len(images))
        train_images = images[:split_index]
        test_images = images[split_index:]
        
        # train 폴더로 복사
        for img in train_images:
            src_img_path = os.path.join(src_label_dir, img)
            dest_img_path = os.path.join(dest_train_label_dir, img)
            shutil.copy(src_img_path, dest_img_path)
        
        # test 폴더로 복사
        for img in test_images:
            src_img_path = os.path.join(src_label_dir, img)
            dest_img_path = os.path.join(dest_test_label_dir, img)
            shutil.copy(src_img_path, dest_img_path)

print("파일 복사가 완료되었습니다.")
```
